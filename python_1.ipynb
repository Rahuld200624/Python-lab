```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("Select which operator")
print("+. addition")
print("-. subtraction")
print("*. multiplication")
print("/. division")

choice = input("Enter your choice (+,-,*,/)")

if choice == "+":
    print("Result: ", num1+num2)
elif choice == "-":
    print("Result: ", num1-num2)
elif choice == "*":
    print("Result: ", num1*num2)
elif choice == "/":
    print("Result: ", num1/num2)

else:
    print("Invalid choice")
        
```
